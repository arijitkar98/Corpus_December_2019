```python
import pickle
import operator
```

```python
with open('count.pckl', 'rb') as f:
    col = pickle.load(f)
```

```python
col = (sorted(col.items(), key = operator.itemgetter(1)))[::-1]
```

```python
just_the_occur = []
just_the_rank = []

entry_num = 1
for entry in col:
    just_the_rank.append(entry_num)
    entry_num += 1
    just_the_occur.append(entry[1])
```

```python
h = []
for u in col:
    if u[1] > 2:
        h.append(u)
```

```python
plt.title("Word Frequencies")
plt.ylabel("Total Number of Occurrences")
plt.loglog(
    just_the_rank,
    just_the_occur,
    basex=10
  )
# plt.plot(
#     just_the_rank,
#     just_the_occur,
#   )
plt.show()
```
